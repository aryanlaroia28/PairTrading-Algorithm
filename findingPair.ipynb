FILTERING OUT STOCKS WHICH HAVE MADE LOSS OVER THE DECADE, SO THAT STRATEGY MAKES PROFIT ON OTHERWISE LOSSY STOCKS

```python
percentage_change = df.pct_change()

total_percentage_change = percentage_change.sum()

loss_stocks = total_percentage_change[total_percentage_change < 0]

print("Stocks that have made a loss from 2013 to 2023:")
print(loss_stocks)
```

FINDING PAIRS WITH CORRELATION>0.9

```python
loss_stocks_data = df[loss_stocks.index]

correlation_matrix_loss_stocks = loss_stocks_data.corr()

threshold = 0.9
high_correlation_pairs = []

for i in range(len(correlation_matrix_loss_stocks.columns)):
    for j in range(i + 1, len(correlation_matrix_loss_stocks.columns)):
        if correlation_matrix_loss_stocks.iloc[i, j] > threshold:
            high_correlation_pairs.append((correlation_matrix_loss_stocks.columns[i], correlation_matrix_loss_stocks.columns[j], correlation_matrix_loss_stocks.iloc[i, j]))

print(f"Number of pairs with correlation > {threshold} among the stocks that made a loss: {len(high_correlation_pairs)}")

for pair in high_correlation_pairs:
    print(f"{pair[0]} and {pair[1]}: {pair[2]:.2f}")
```

FINDING COINTEGRATED PAIRS TO ASSURE MEAN REVERSION

```python
import statsmodels.api as sm
from statsmodels.tsa.stattools import coint

def check_cointegration(stock1, stock2):
    stock1_prices = df[stock1].dropna()
    stock2_prices = df[stock2].dropna()
    min_length = min(len(stock1_prices), len(stock2_prices))
    stock1_prices = stock1_prices[-min_length:]
    stock2_prices = stock2_prices[-min_length:]
    score, p_value, _ = coint(stock1_prices, stock2_prices)
    return p_value

cointegration_threshold = 0.05
cointegrated_pairs = []

for pair in high_correlation_pairs:
    stock1, stock2, correlation = pair
    p_value = check_cointegration(stock1, stock2)
    if p_value < cointegration_threshold:
        cointegrated_pairs.append((stock1, stock2, p_value))

        

for pair in cointegrated_pairs:
    print(f"{pair[0]} and {pair[1]}: p-value = {pair[2]:.4f}")
```

PLOTTING THE CHOOSEN STOCK PAIR

```python
import yfinance as yf
import matplotlib.pyplot as plt

def fetch_stock_data(ticker, start, end):
    stock_data = yf.download(ticker, start=start, end=end)
    return stock_data['Adj Close']

def plot_cointegrated_pairs(pair, stock1_prices, stock2_prices):
    stock1, stock2, _ = pair
    
    plt.figure(figsize=(10, 6))
    plt.plot(stock1_prices, label=stock1)
    plt.plot(stock2_prices, label=stock2)
    plt.title(f"{stock1} and {stock2} - Cointegrated Pair")
    plt.xlabel("Time")
    plt.ylabel("Price")
    plt.legend()
    plt.show()

for pair in cointegrated_pairs:
    print(f"{pair[0]} and {pair[1]}: p-value = {pair[2]:.4f}")
    stock1_prices = fetch_stock_data(pair[0], "2018-01-01", "2023-01-01")
    stock2_prices = fetch_stock_data(pair[1], "2018-01-01", "2023-01-01")
    plot_cointegrated_pairs(pair, stock1_prices, stock2_prices)
```

PLOTTING THE RATIO OF STOCK PAIR AND THEIR MEAN

```python
import numpy as np
import matplotlib.pyplot as plt


def calculate_ratio(stock1, stock2):
    ratio = df[stock2].dropna() / df[stock1].dropna()
    return ratio

for pair in cointegrated_pairs:
    stock1, stock2, p_value = pair
    ratio = calculate_ratio(stock1, stock2)
  
    mean_ratio = np.mean(ratio)
    
    plt.figure(figsize=(12, 6))
    plt.plot(ratio, label='Ratio')
    plt.axhline(mean_ratio, color='red', linestyle='--', linewidth=2, label='Mean of Ratio')
    plt.title(f'Ratio between {stock1} and {stock2} (p-value: {p_value:.4f})')
    plt.xlabel('Date')
    plt.ylabel('Ratio')
    plt.legend()
    plt.show()
```
